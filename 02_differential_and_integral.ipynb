```python
# 2차원 함수 plot(2차 함수)
%matplotlib inline

def f2(x, y):
    
    return (1-x)**2 + 100.0 * (y - x**2)**2


x = np.linspace(-4, 4, 800)
y = np.linspace(-3, 3, 600)

X, Y = np.meshgrid(x, y)

Z = f2(X,Y)
print("Z :")
print(Z, "\n")

fig = plt.figure(figsize = (12, 12))
ax1 = fig.add_subplot(111, projection = "3d")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_ylabel("z")

ax1.plot_surface(X,Y,Z)
```

```python
from random import *

a = randint(1, 100)
b = randint(1, 100)
c = randint(1, 100)
d = randint(1, 100)
e = randint(1, 100)
f = randint(1, 100)

test_matrix = np.array([[a, b],
                       [c, d],
                       [e, f]])

result = test_matrix @ np.linalg.inv(test_matrix.T @ test_matrix) @ test_matrix.T

print("result_matrix : ", "\n")
print(result, "\n")
print("result_matrix_squared : ", "\n")
print(result @ result, "\n")
print("위 두 행렬을 통해, result 행렬이 멱등행렬임을 알 수 있음", "\n")
print("*"*80, "\n")

print("result 멱등행렬의 역행렬 : ", "\n")
print(np.linalg.inv(result))

```

```python
# scipy의 서브패키지 optimize 내 minimize 함수로 최적화 실시
def f1(x):
    
    return (x - 2) **2 + 2

start = 0
result = sp.optimize.minimize(f1, start)
print(result)
```

```python
# 그래디언트 벡터값을 반환하는 함수를 인자로 넣어서, minimize함수 호출
def f1g(x):
    
    return 2 * (x -2)

start = np.array([0])
result = sp.optimize.minimize(f1, start, jac = f1g)
print(result)
```

```python
# 다변수함수 최적화하는 방법
# f2함수는 벡터를 입력 받음
def f2(x):
    
    return (1 - x[0])**2 + 100 * (x[1] -x[0]**2)**2

start = (-2, -2)
result = sp.optimize.minimize(f2, start)
print(result)

```

```python
# 2차원 로젠브록 함수 최적화 실시
def f3(x):
    
    return (1 - x[0])**2 + 100 * (x[1] - x[0]**2)**2

start = np.array([2, 0.3])
result = sp.optimize.minimize(f3, start)
print(result)
```

```python
# 로젠브록 함수의 그래디언트 벡터를 minimize의 인수로 입력한 후, 최적화 실시
def f3g(x):
    
    return np.array([2 * (x[0] - 1) - 400 * x[0] * (x[1] - x[0]**2), 200 * (x[1] - x[0]**2)])

start = np.array([2, 0.3])
result = sp.optimize.minimize(f3, start, jac=f3g)
print(result)    
```

```python
# scipy를 사용하여 등식 제한조건이 있는 최적화 문제 계산하기
def f1(x):
    
    return x[0] ** 2 + x[1] ** 2

def eq_constraint(x):
    
    return x[0] + x[1] - 1

sp.optimize.fmin_slsqp(f1, np.array([1,1]), eqcons=[eq_constraint])
```

```python
'''부등식 제한조건이 있는 최적화 문제 풀이'''
%matplotlib inline

# 목적함수 : x1^2 + x2^2
def f1(x1, x2):
    
    return x1 ** 2 + x2 ** 2

# 해당 목적함수 3d plot
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)

# 제한조건 : x1 + x2 = 0
x2_g = 1 - x1

Y = f1(X1, X2)

fig1 = plt.figure(figsize=(7, 7))
ax1 = fig1.add_subplot(1, 1, 1)
ax1 = fig1.gca(projection = "3d")
ax1.plot_surface(X1, X2, Y)
ax1.set_xlabel("$x1$")
ax1.set_ylabel("$x2$")
ax1.set_title("$x1$^2 + $x2$^2함수 그리기")

fig2 = plt.figure(figsize=(7, 7))
ax2 = fig2.add_subplot(1, 1, 1)
ax2.contour(X1, X2, Y, colors = "gray", levels = [0.5, 2, 8])
ax2.set_xlabel("$x1$")
ax2.set_ylabel("$x2$")

plt.show()
```
