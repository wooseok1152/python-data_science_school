```python
import pandas_datareader.data as pdr

# yahoo finance에서 받아오는 주가 데이터의 '종가'는 '수정종가'임
# '대한제당(001790)'은 '무상증자 100%'실시로 인해 4월 30일에 권리락이 발생하였음
# 밑에 제시된 해당 종목 주가이력의 'Close(종가)'컬럼에는 해당 무상증자가 고려된 '수정종가'가 입력되어 있음
print("* 대한제당 주가내역 :")
display(pdr.get_data_yahoo('001790.KS', start = '2021-04-15', end = '2021-05-15'))

# 코스피 종목들에 대해선 전일 주가내역만 받아올 수 있음
print("\n", "* 카카오게임즈 주가내역 :")
display(pdr.get_data_yahoo('293490.KQ'))
```
