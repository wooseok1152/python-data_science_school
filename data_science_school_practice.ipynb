```python
'''boston 집값 dataset의 4개 데이터를 활용하여, 회귀식의 회귀계수 구하는 방법'''

from sklearn.datasets import load_boston
boston = load_boston()
x = boston.data
y = boston.target

# CRIM(범죄율), NOX(공기 오염도), RM(방 개수), AGE(오래된 정도)에 대한 4개의 데이터 레코드, 즉, 미지수 4개
A = x[:4, [0, 4, 5, 6]]

# 각 데이터 레코드에 대한 결과값(주택 가격) 4개, 즉 4개의 방정식
b = y[:4]

print(A, "\n")
print(b, "\n")

# 회기계수 구하는 방법 : 특징 행렬의 역행렬과 종속 변수에 대한 열벡터를 dot product실시
coefficient_vector = np.linalg.inv(A) @ b
print(coefficient_vector)
```

```python
'''pgmpy의 JPD를 활용하여 결합확률분포 나타내기'''
from pgmpy.factors.discrete import JointProbabilityDistribution as JPD

# sex('0'이면 남자, '1'이면 여자) 한 개의 확률변수에 대한 확률분포 나타내기
p_sex = JPD(["sex"], [2], np.array([12, 8])/20)
print(p_sex)
```

```python
import scipy as sp
import scipy.stats
```

```python
# 정규분포 객체 생성
# loc : 일반적으로 분포의 기댓값
# scale : 일반적으로 분포의 표준편차

rv = sp.stats.norm(loc = 1, scale = 2)
```

```python
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)

plt.plot(xx, pdf)
plt.title("정규분포 확률밀도함수")
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

```python
# 기댓값이 0이고 표준편차가 0.1인 정규분포의 객체를 만들고 확률밀도함수 그리기
rv = sp.stats.norm(loc = 0, scale = 0.1)
xx = np.linspace(-1, 1, 1000)
pdf = rv.pdf(xx)

plt.plot(xx, pdf)
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.title('기댓값이 0이고 표준편차가 0.1인 정규분포')
plt.show()

# 해당 확률밀도의 최댓값 구하기
print("최댓값 :", rv.pdf(0))
```

```python
# 누적분포함수 그리기
rv = sp.stats.norm(loc = 1, scale = 2)
xx = np.linspace(-8, 8,100)
cdf = rv.cdf(xx)

plt.plot(xx, cdf)
plt.title("누적분포함수")
plt.xlabel('$x$')
plt.ylabel('F(x)')
plt.show()
```

```python
# 특정 분포에 기반한 무작위 표본을 생성하는 법
# size : 표본 생성 시 생성될 표본 크기
# random_state : 표본 생성 시 사용되는 시드값

rv.rvs(size = (3, 5), random_state = 0)
```

```python
# 변환 확률변수의 시뮬레이션

%matplotlib inline

rv1 = sp.stats.uniform()
rv2 = sp.stats.uniform()

np.random.seed(0)
x_1 = rv1.rvs(size = 50000)
x_2 = rv2.rvs(size = 50000)
x_3 = x_1 + x_2

plt.figure(figsize = (12, 5))

plt.subplot(131)
sns.distplot(x_1, kde=False)
plt.title("균일분포")
plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.subplot(132)
sns.distplot(x_2, kde=False)
plt.title("균일분포")
plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.subplot(133)
sns.distplot(x_3, kde=False)
plt.title("균일분포 표본의 합의 분포")
plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.show()
```

```python
# 균일분포 확률분포에서 두 개가 아닌 10개 표본값을 생성하여, 그 값의 합을 구하면 어떤 모양의 분포를 이루는지 파악

%matplotlib inline

rv1 = sp.stats.uniform()
rvs_sum_list = []

for i in range(1000) :

    rvs = rv1.rvs(10)
    rvs_sum = np.sum(rvs)
    rvs_sum_list.append(rvs_sum)
    
plt.figure()
sns.distplot(rvs_sum_list, kde = False)
plt.xlabel("rvs_sum")
plt.ylabel("relative frequency")
plt.show()

```

```python
from tabulate import tabulate

# 베르누이 확률변수의 시뮬레이션

mu = 0.6
rv = sp.stats.bernoulli(mu)
print("* rv typet :", type(rv), "\n")

xx = [0, 1]
plt.bar(xx, rv.pmf(xx))
plt.xlim(-1, 2)
plt.ylim(0, 1)

# 눈금 표시 사용자 설정하는 법
plt.xticks([0, 1], ['x=0', 'y=0'])

plt.xlabel('표본값')
plt.ylabel('P(x)')
plt.title('베르누이 확률변수의 확률질량함수')
plt.show()

x = rv.rvs(100, random_state = 0)
print('* 베르누이 확률변수 시뮬레이션 결과 :', "\n", "\n", x)

plt.figure()
sns.countplot(x)
plt.title('베르누이 확률변수의 시뮬레이션 결과')
plt.xlabel('표본값')
plt.show()

print('* 이론과 시뮬레이션 비교', "\n")
y = np.bincount(x) / len(x) 
df = pd.DataFrame({"이론" : rv.pmf(xx), "시뮬레이션" : y})
print(tabulate(df, headers = 'keys', tablefmt = 'psql'))

df2 = df.stack().reset_index()
df2.columns = ['표본값', '유형', '비율']
plt.figure()
sns.barplot(x = '표본값', y = '비율', hue = '유형', data = df2)
plt.title('베르누이분포의 이론적 분포와 시뮬레이션 분포')
plt.show()
```

```python
# 주식 수익률 분포(정규분포)

%matplotlib inline

import pandas_datareader.data as pdr

print("* 나스닥 지수 이력 데이터프레임 확인 :")
data = pdr.get_data_fred('NASDAQCOM', start = '2009-01-01', end = '2018-12-31')
display(data[:10])

print("* NULL레코드 제거")
data = data["NASDAQCOM"]
data = data.dropna()     # NULL 레코드 제거
display(data)

plt.plot(data)
plt.title("나스닥 지수")
plt.show()

daily_returns = data.pct_change().dropna()
mean = daily_returns.mean()
std = daily_returns.std()
print("* 평균 일간수익률 : {:.2f}%".format(mean * 100))
print("* 평균 일간수익률 : {:.2f}%".format(std * 100))
```

```python
# 로그정규분포

import scipy as sp
import scipy.stats

mu = 1
rv = sp.stats.norm(loc = mu)
x1 = rv.rvs(1000, random_state = 0)
print("* 'x1'내 10개 원소 확인 :")
print(x1[:10], "\n")

s = 0.5
x2 = np.exp(s * x1)
print("* 'x2'내 10개 원소 확인 :")
print(x2[:10], "\n")

plt.figure()

plt.subplot(121)
sns.distplot(x1, kde=False)
plt.title("정규분포")

plt.subplot(122)
sns.distplot(x2, kde=False)
plt.title("로그정규분포")

plt.tight_layout()
plt.show()
```

```python
# Q-Q plot을 통한 정규성 검정

import scipy as sp
import scipy.stats
from scipy.stats.morestats import _calc_uniform_order_statistic_medians

%matplotlib inline
mpl.rc('font', family = 'Apple SD Gothic Neo')

mu = 0
std = 1
rv = sp.stats.norm(mu, std)

x = rv.rvs(20, random_state = 0)


x_sorted = np.sort(x)
print("x_sorted :")
print(x_sorted, "\n")

position = _calc_uniform_order_statistic_medians(len(x))
print("position :")
print(position, "\n")

qf = rv.ppf(position)
print("qf :")
print(qf ,"\n")

plt.figure()
plt.scatter(qf, x_sorted)
plt.title("Q-Q plot")
plt.xlabel("이론적인 위칫값")
plt.ylabel("정렬된 표본 데이터")
plt.axis("equal")
plt.show()

plt.figure()
sp.stats.probplot(x, plot = plt)
plt.axis("equal")
plt.show()
```
