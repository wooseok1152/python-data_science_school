```python
'''boston 집값 dataset의 4개 데이터를 활용하여, 회귀식의 회귀계수 구하는 방법'''

from sklearn.datasets import load_boston
boston = load_boston()
x = boston.data
y = boston.target

# CRIM(범죄율), NOX(공기 오염도), RM(방 개수), AGE(오래된 정도)에 대한 4개의 데이터 레코드, 즉, 미지수 4개
A = x[:4, [0, 4, 5, 6]]

# 각 데이터 레코드에 대한 결과값(주택 가격) 4개, 즉 4개의 방정식
b = y[:4]

print(A, "\n")
print(b, "\n")

# 회기계수 구하는 방법 : 특징 행렬의 역행렬과 종속 변수에 대한 열벡터를 dot product실시
coefficient_vector = np.linalg.inv(A) @ b
print(coefficient_vector)
```

```python
from random import *

a = randint(1, 100)
b = randint(1, 100)
c = randint(1, 100)
d = randint(1, 100)
e = randint(1, 100)
f = randint(1, 100)

test_matrix = np.array([[a, b],
                       [c, d],
                       [e, f]])

result = test_matrix @ np.linalg.inv(test_matrix.T @ test_matrix) @ test_matrix.T

print("result_matrix : ", "\n")
print(result, "\n")
print("result_matrix_squared : ", "\n")
print(result @ result, "\n")
print("위 두 행렬을 통해, result 행렬이 멱등행렬임을 알 수 있음", "\n")
print("*"*80, "\n")

print("result 멱등행렬의 역행렬 : ", "\n")
print(np.linalg.inv(result))

```

```python
# scipy의 서브패키지 optimize 내 minimize 함수로 최적화 실시
def f1(x):
    
    return (x - 2) **2 + 2

start = 0
result = sp.optimize.minimize(f1, start)
print(result)
```

```python
# 그래디언트 벡터값을 반환하는 함수를 인자로 넣어서, minimize함수 호출
def f1g(x):
    
    return 2 * (x -2)

start = np.array([0])
result = sp.optimize.minimize(f1, start, jac = f1g)
print(result)
```

```python
# 2차원 로젠브록 함수 최적화 실시
def f3(x):
    
    return (1 - x[0])**2 + 100 * (x[1] - x[0]**2)**2

start = np.array([2, 0.3])
result = sp.optimize.minimize(f3, start)
print(result)
```

```python
# 로젠브록 함수의 그래디언트 벡터를 minimize의 인수로 입력한 후, 최적화 실시
def f3g(x):
    
    return np.array([2 * (x[0] - 1) - 400 * x[0] * (x[1] - x[0]**2), 200 * (x[1] - x[0]**2)])

start = np.array([2, 0.3])
result = sp.optimize.minimize(f3, start, jac=f3g)
print(result)    
```

```python
'''부등식 제한조건이 있는 최적화 문제 풀이'''
%matplotlib inline

# 목적함수 : x1^2 + x2^2
def f1(x1, x2):
    
    return x1 ** 2 + x2 ** 2

# 해당 목적함수 3d plot
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)

# 제한조건 : x1 + x2 = 0
x2_g = 1 - x1

Y = f1(X1, X2)

fig1 = plt.figure(figsize=(7, 7))
ax1 = fig1.add_subplot(1, 1, 1)
ax1 = fig1.gca(projection = "3d")
ax1.plot_surface(X1, X2, Y)
ax1.set_xlabel("$x1$")
ax1.set_ylabel("$x2$")
ax1.set_title("$x1$^2 + $x2$^2함수 그리기")

fig2 = plt.figure(figsize=(7, 7))
ax2 = fig2.add_subplot(1, 1, 1)
ax2.contour(X1, X2, Y, colors = "gray", levels = [0.5, 2, 8])
ax2.set_xlabel("$x1$")
ax2.set_ylabel("$x2$")

plt.show()
```

```python
'''사건과 각 사건에 대한 확률을 표현하는 방법'''

# 사건
A = frozenset([])
B = frozenset(["H"])
C = frozenset(["T"])
D = frozenset(["H", "T"])

# 확률
P = {A : 0, B : 0.4, C : 0.6, D : 1}
P
```

```python
'''pgmpy의 JPD를 활용하여 결합확률분포 나타내기'''
from pgmpy.factors.discrete import JointProbabilityDistribution as JPD

# sex('0'이면 남자, '1'이면 여자) 한 개의 확률변수에 대한 확률분포 나타내기
p_sex = JPD(["sex"], [2], np.array([12, 8])/20)
print(p_sex)
```

```python
# sex와 머리 길이('0'이면 긴 머리, '1'이면 짧은 머리) 두 개의 확률변수에 대한 결합확률분포 나타내기
p_sex_hair = JPD(['sex', 'hair_len'], [2, 2], np.array([3, 9, 7, 1, ])/20)
print(p_sex_hair)
```

```python
# 위 결합확률분포를 활용하여, sex 확률변수에 대한 확률분포 나타내기
p_sex = p_sex_hair.marginal_distribution(["sex"], inplace=False)
print(p_sex)
```

```python
# 위 결합확률분포를 활용하여, hair_len 확률변수에 대한 확률분포 나타내기
p_hair = p_sex_hair.marginal_distribution(["hair_len"], inplace=False)
print(p_hair)
```

```python
# 위 결합확률분포를 활용하여, 두 확률변수의 독립여부 확인
p_sex_hair.check_independence(["sex"], ["hair_len"])
```

```python
'''분포가 비대칭일 때, 표본평균,표본중앙값, 표본최빈값의 차이'''
np.random.seed(1)
x = np.random.normal(size = 1000)
x = np.hstack([x, 5 * np.ones(50)])
# print(x)

bins = np.linspace(-6, 6, 12*4 + 1)
# print(bins, len(bins))

ns, _ = np.histogram(x, bins = bins)
# print(ns, "\n")
# print(_)

sample_mean = np.mean(x)
sample_median = np.median(x)

# 'ns' 벡터 내 최대값 요소의 index
mode_index = np.argmax(ns)
sample_mode = 0.5 * (bins[mode_index] + bins[mode_index + 1])

plt.figure(figsize=(10,6))
sns.distplot(x, bins=bins)

# 'c' 매개변수 : verticle line의 color를 의미, "ls" 매개변수 : line style을 의미
plt.axvline(sample_mean, c = "k", ls=":", label = "표본평균")
plt.axvline(sample_median, c = "k", ls = "--", label = "표본중앙값")
plt.axvline(sample_mode, c = "k", ls = "-", label = "표본최빈값")
plt.title("분포가 비대칭일 때, 표본평균, 표본중앙값, 표본최빈값의 차이")
plt.legend()
plt.xlabel("x")
plt.show()
```

# 확률분포

```python
import scipy as sp
import scipy.stats
```

```python
# 정규분포 객체 생성
# loc : 일반적으로 분포의 기댓값
# scale : 일반적으로 분포의 표준편차

rv = sp.stats.norm(loc = 1, scale = 2)
```

```python
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)

plt.plot(xx, pdf)
plt.title("정규분포 확률밀도함수")
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

```python
# 기댓값이 0이고 표준편차가 0.1인 정규분포의 객체를 만들고 확률밀도함수 그리기
rv = sp.stats.norm(loc = 0, scale = 0.1)
xx = np.linspace(-1, 1, 1000)
pdf = rv.pdf(xx)

plt.plot(xx, pdf)
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.title('기댓값이 0이고 표준편차가 0.1인 정규분포')
plt.show()

# 해당 확률밀도의 최댓값 구하기
print("최댓값 :", rv.pdf(0))
```

```python
# 누적분포함수 그리기
rv = sp.stats.norm(loc = 1, scale = 2)
xx = np.linspace(-8, 8,100)
cdf = rv.cdf(xx)

plt.plot(xx, cdf)
plt.title("누적분포함수")
plt.xlabel('$x$')
plt.ylabel('F(x)')
plt.show()
```

```python
# 특정 분포에 기반한 무작위 표본을 생성하는 법
# size : 표본 생성 시 생성될 표본 크기
# random_state : 표본 생성 시 사용되는 시드값

rv.rvs(size = (3, 5), random_state = 0)
```

```python
# 랜덤 표본 생성 결과
rv = sp.stats.norm(loc = 1, scale = 2)

sns.distplot(rv.rvs(size = 10000, random_state = 0))
plt.title('.랜덤 표본 생성 결과')
plt.xlabel('표본값')
plt.ylabel('count')
plt.xlim(-8 , 8)
plt.show()
```
