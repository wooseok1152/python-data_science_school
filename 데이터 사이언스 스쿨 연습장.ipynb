```python
%matplotlib inline

def f2(x, y):
    
    return (1-x)**2 + 100.0 * (y - x**2)**2


x = np.linspace(-4, 4, 800)
y = np.linspace(-3, 3, 600)

X, Y = np.meshgrid(x, y)

Z = f2(X,Y)
print(Z, "\n")

fig = plt.figure(figsize = (12, 12))
ax1 = fig.add_subplot(111, projection = "3d")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax1.plot_surface(X,Y,Z)

# fig2 = plt.figure(figsize = (12, 12))
# ax2 = fig2.add_subplot(111)
# CS = ax2.contour(X, Y, Z)
```

```python
from random import *

a = randint(1, 100)
b = randint(1, 100)
c = randint(1, 100)
d = randint(1, 100)
e = randint(1, 100)
f = randint(1, 100)

test_matrix = np.array([[a, b],
                       [c, d],
                       [e, f]])

result = test_matrix @ np.linalg.inv(test_matrix.T @ test_matrix) @ test_matrix.T

print("result_matrix : ", "\n")
print(result, "\n")
print("result_matrix_squared : ", "\n")
print(result @ result, "\n")
print("위 두 행렬을 통해, result 행렬이 멱등행렬임을 알 수 있음", "\n")
print("*"*80, "\n")

print("result 멱등행렬의 역행렬 : ", "\n")
print(np.linalg.inv(result))

```

```python
x = np.array([[1], [2], [3], [4]])
print(x, "\n")

X = x @ x.T
print(X, "\n")

vector_multiply = x.T @ x
print(vector_multiply, "\n")

H = np.eye(4) - (2*X/vector_multiply)
print(H @ H)
```
